```python
# Load Libraries
import json
import pandas as pd
import geopandas as gpd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

# Load project configuration
with open('config.json', 'r') as f:
    config = json.load(f)

with open(Path(config['paths']['data_processed']) / 'analysis_config.json', 'r') as f:
    analysis_config = json.load(f)

# Extract paths
data_raw = Path(config['paths']['data_raw'])
data_processed = Path(config['paths']['data_processed'])
climate_path = Path(config['paths']['climate'])
demographics_path = Path(config['paths']['demographics'])
infrastructure_path = Path(config['paths']['infrastructure'])
shapefiles_path = Path(config['paths']['shapefiles'])
redlining_path = Path(config['paths']['redlining'])
```

```python
# Load neighborhood data from preprocessing
neighborhoods = gpd.read_file(data_processed / 'neighborhoods_clean.gpkg')

# Load climate data
climate_data = pd.read_csv(climate_path / 'nyc_daily_climate.csv')

# Load demographic data
demographics_data = pd.read_csv(demographics_path / 'nyc_demographics.csv')
```

```python
# Load infrastructure data
cooling_centers = pd.read_csv(infrastructure_path / 'nyc_cooling_centers.csv')
green_spaces = pd.read_csv(infrastructure_path / 'nyc_green_spaces.csv')
urban_design = pd.read_csv(infrastructure_path / 'nyc_urban_design.csv')
```

```python
# Save all processed datasets for analysis
neighborhoods.to_file(data_processed / 'neighborhoods_final.gpkg', driver='GPKG')
climate_data.to_csv(data_processed / 'climate_final.csv', index=False)
demographics_data.to_csv(data_processed / 'demographics_final.csv', index=False)
cooling_centers.to_csv(data_processed / 'cooling_centers_final.csv', index=False)
green_spaces.to_csv(data_processed / 'green_spaces_final.csv', index=False)
urban_design.to_csv(data_processed / 'urban_design_final.csv', index=False)
```
